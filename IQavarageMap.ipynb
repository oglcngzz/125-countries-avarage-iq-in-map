```python
world_map = folium.Map(location=[0, 0], zoom_start=2)

def color_producer(iq):
    if iq < 90:
        return 'red'
    elif 90 <= iq < 100:
        return 'orange'
    else:
        return 'green'

for index, row in df_cleaned.iterrows():
    popup_html = f"<strong>Country:</strong> {row['Country']}<br><strong>IQ:</strong> {row['Average IQ']}"
    
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        tooltip=row['Country'],
        popup=folium.Popup(popup_html, max_width=200),
        icon=folium.Icon(color=color_producer(row['Average IQ']))
    ).add_to(world_map)

world_map.save('world_map.html')
```
